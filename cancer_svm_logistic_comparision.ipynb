```python
# libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
data= pd.read_csv(r"C:\Users\Lenovo\OneDrive\Desktop\py\data\breast\data.csv")
print(data.shape)
data.head()
```

```python
data.drop('id',axis=1,inplace=True)
data.head()
```

```python
# data processing
nulls =data.isnull().sum()
print(nulls)
```

```python
#data.drop('Unnamed: 32',axis=1,inplace=True)
data.head()
```

```python
# M: have cancer
cis212 =data['diagnosis'].value_counts()
print(cis212)
```

```python
onc = OneHotEncoder()
A =onc.fit_transform(data[['diagnosis']]).toarray()[:,1]
ndf = pd.DataFrame(A)
ndf.value_counts()

data.drop('diagnosis',axis=1,inplace=True)
data.head()
```

```python
data = pd.concat([data,ndf],axis=1)
data.rename(columns={0: 'diagnosis'}, inplace=True)
data.head()
```

```python
x= data.drop('diagnosis',axis=1)
x.head()
```

```python
y=data['diagnosis']
y.value_counts()
```

```python
# splitting
accuracy_l=[]
accuracy_s=[]
test_sizes =[]

for i in [0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45]:
    xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=i, random_state=9)
    # scaling
    sc = StandardScaler()
    sc.fit(xtrain)
    xtrain = sc.transform(xtrain)
    xtest= sc.transform(xtest)

    
    # Logistic regression
    model = LogisticRegression()
    model.fit(xtrain,ytrain)
    pred = model.predict(xtest)
    acclr = accuracy_score(ytest, pred)
    accuracy_l.append(acclr)


    #SVM
    svc = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=9)
    svc.fit(xtrain, ytrain)
    pred = svc.predict(xtest)
    accrbf = accuracy_score(ytest, pred)
    accuracy_s.append(accrbf)
    
    
    test_sizes.append(i)


lm = max(accuracy_l)
sm = max(accuracy_s)

if lm>sm:
    print("Logistic")
elif sm>lm:
    print('SVM')

plt.plot(test_sizes, accuracy_l, label='Logistic Regression')
plt.plot(test_sizes, accuracy_s, label='SVM (RBF)')
plt.xlabel("Test Size")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
gamma_values = []

accuracy_svm = []

for g in range(1,300,5):
    svc = SVC(kernel='rbf', C=1.0, gamma=(g/10000), random_state=9)
    svc.fit(xtrain, ytrain)
    pred = svc.predict(xtest)
    acc = accuracy_score(ytest, pred)
    accuracy_svm.append(acc)
    gamma_values.append(g)

# Find best gamma
best_acc = max(accuracy_svm)
best_gamma = gamma_values[accuracy_svm.index(best_acc)]
print(f"Best Gamma: {best_gamma} with Accuracy: {best_acc:.4f}")

# Plot
plt.plot(gamma_values, accuracy_svm, marker='o')
plt.xscale('log')  # gamma varies exponentially, so log scale looks better
plt.xlabel("Gamma")
plt.ylabel("Accuracy")
plt.title("SVM Accuracy vs Gamma")
plt.show()
```
